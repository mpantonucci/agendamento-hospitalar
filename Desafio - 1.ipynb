```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
mes = "2020/Ago"
coluna = "Unidade da Federação"

dados = pd.read_csv("/content/basedados.csv", encoding="ISO-8859-1", skiprows=3, sep=";", skipfooter=12, thousands=".", decimal=",")
dados[mes] = pd.to_numeric(dados[mes], errors='coerce')
dados[coluna] = dados[coluna].str.slice(start=3)
dados = dados.sort_values(by=[mes], ascending=True)
ax = dados.plot(x=coluna, y=mes, figsize=(15,10), kind="bar", legend=None)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Gastos médicos por UF - Agosto/2020")
plt.show()


```
